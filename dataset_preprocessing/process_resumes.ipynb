```python
import os
import pandas as pd

# construct a csv file with the path to the files in the dataset and the professional area as the folder name.
# the folder with the subfolders and the resumes is in ../data/resumes/...

# Define the path to the dataset
dataset_path = "../data/resumes/"

# Initialize an empty list to store the data
data = []

# Walk through the dataset directory
for professional_area in os.listdir(dataset_path):
    area_path = os.path.join(dataset_path, professional_area)
    if os.path.isdir(area_path):  # Ensure it's a directory
        for file in os.listdir(area_path):
            if file.endswith(".pdf"):  # Only consider PDF files
                resume_path = os.path.join(area_path, file)
                data.append({"professional_area": professional_area, "resume_path": resume_path})

# Create a DataFrame from the collected data
df = pd.DataFrame(data)

# Print the DataFrame to check the contents
print(df.head())
print(df.tail())
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each professional area
area_counts = df['professional_area'].value_counts()

# Plot the distribution as a bar chart
area_counts.plot(kind='bar', figsize=(15, 6), color='skyblue')

# Add labels and title
plt.xlabel('Professional Area')
plt.ylabel('Number of Resumes')
plt.title('Distribution of Professional Areas')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
